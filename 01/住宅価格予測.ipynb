```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# データの読み込み
train_data = pd.read_csv('./data/train.csv')
test_data = pd.read_csv('./data/test.csv')
submit_data = pd.read_csv("./data/sample_submission.csv")
```

```python
train_data.columns
```

```python
# データ数の把握
print("{} rows and {} features in train set".format(train_data.shape[0], train_data.shape[1]))
print("{} rows and {} features in test set".format(test_data.shape[0], test_data.shape[1]))
print("{} rows and {} features in submit set".format(submit_data.shape[0], submit_data.shape[1]))
```

```python
# 訓練データにどれくらいの欠損があるのか
train_data.isnull().sum()
```

```python
# 欠損値の可視化
# 欠損値に規則性があるのかを見る
plt.figure(figsize=(18,9))
sns.heatmap(train_data.isnull(), cbar=False)
```

```python
train_data.describe()
```

```python
# 分類では使える。　これ価格だから、横軸がめっちゃ出てくる
target_col = "SalePrice"
sns.countplot(x=target_col, data=train_data)
```

```python
# 相関関係
# 特に価格との相関があるパラメータは無さそう
train_data.corr()
```

```python
# 分類では使える。
# 価格じゃなくて別のパラメータならどうか
# RL多め
target_col = "MSZoning"
sns.countplot(x=target_col, data=train_data)
```

```python
# 他のパラメータも見てみる
plt.figure(figsize=(5,5))
target_col = "MSSubClass"
sns.countplot(x=target_col, data=train_data)
plt.figure(figsize=(5,5))
target_col = "OverallQual"
sns.countplot(x=target_col, data=train_data)
```

```python
# 価格をプロットする良い感じのコードあったのでコピペ

# 売却価格のヒストグラム
sns.distplot(train_data['SalePrice'])
# 売却価格の概要をみてみる
print(train_data["SalePrice"].describe())
print(f"歪度: {round(train_data['SalePrice'].skew(),4)}" )
print(f"尖度: {round(train_data['SalePrice'].kurt(),4)}" )
```

```python
# 歪度(わいど)
# 分布を正規分布と比較したときの偏り度合い（非対称性）を記す尺度。0に近いほど左右対称
# 尖度(せんど)
# 分布を正規分布と比較した時の尖り具合を表す尺度。0に近いほど正規分布に近い。また、性の大きい値を取るほど、ピークふきんに標本が集中しており、負の大きい値をとるほど分布が散らばっているt
```
